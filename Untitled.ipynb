```python
# Extract a variable from the file
data = ds["air"].data

# Plot the image using matplotlib
plt.imshow(data[0], origin='lower')
plt.colorbar()
plt.title("Schneider Cut")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
